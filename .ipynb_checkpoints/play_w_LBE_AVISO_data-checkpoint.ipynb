## Plot stuff! 

```python
# Params to make plots pretty
fontsize = 18
plt.rc('font', size=fontsize)          # controls default text sizes
plt.rc('axes', titlesize=fontsize)     # fontsize of the axes title
plt.rc('axes', labelsize=fontsize)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=fontsize)    # fontsize of the tick labels
plt.rc('ytick', labelsize=fontsize)    # fontsize of the tick labels
plt.rc('legend', fontsize=fontsize)    # legend fontsize
plt.rc('figure', titlesize=fontsize)  # fontsize of the figure title
```

Timeline of LBE track IDs in the AVISO dataset

```python
fig,ax = plt.subplots(1,1,figsize=(12,5))

for t in np.unique(ds.track):
    times = np.array(ds.time[np.where(ds.track == t)[0]])
    ax.plot(times, np.ones_like(times,dtype=int) * t, 'o', label=f'Track {t}')

ax.set_xlabel('Date')
ax.set_ylabel('Track')
ax.set_title('Timeline of Tracks')
ax.legend(bbox_to_anchor=(1,1))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Plot the eddy on any given day. Date format: 'YYYY-MM-DD'

```python
date_to_plot = '2020-05-08'
ARGO_lon1,ARGO_lat1 = 4,69 # Example argo location
ARGO_lon2,ARGO_lat2 = 3,69

############################
contour_lons,contour_lats,center_lon,center_lat,track_ID = get_LBE(ds,date_to_plot)

# Plot
plt.plot(contour_lons,contour_lats)
plt.scatter(center_lon,center_lat,marker='*',s=100)
plt.scatter(ARGO_lon1,ARGO_lat1,c='r',s=100,label='Argo in eddy')
plt.scatter(ARGO_lon2,ARGO_lat2,c='orange',s=100,label='Argo out of eddy')

# Labels
plt.title('LBE on %s'%(date_to_plot))
plt.xlabel('Longitude ($^{\circ}$E)');
plt.ylabel('Latitude ($^{\circ}$N)');
plt.legend(bbox_to_anchor=(1,1))
```

in_eddy() function can be used to determine if a (lat,lon,date) point is in the eddy or not

```python
print('Red is in the eddy: %s'%(in_eddy(ds,ARGO_lat1,ARGO_lon1,date_to_plot)))
print('Orange is in the eddy: %s'%(in_eddy(ds,ARGO_lat2,ARGO_lon2,date_to_plot)))
```

## LBE major splits / merges

```python
def index_cmap(NUM_COLORS,cmap):
    cm = pylab.get_cmap(cmap)
    colors = [cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)]
    return colors
```

```python
def plot_LBE_overlap_dates(id1, id2):

    # GRAB OVERLAP DATES
    if (id1 == 667049) or (id2 == 667049):
        overlap_dates = ['2020-04-29','2020-04-30','2020-05-01','2020-05-02','2020-05-03','2020-05-04','2020-05-05','2020-05-06','2020-05-07',
                '2020-05-08','2020-05-09','2020-05-10','2020-05-11','2020-05-12','2020-05-13','2020-05-14','2020-05-15','2020-05-16',
                '2020-05-17','2020-05-18','2020-05-19','2020-05-20','2020-05-21']
    elif (id1 == 586453) or (id2 == 586453):
        overlap_dates = ['2014-09-14','2014-09-15','2014-09-16','2014-09-17','2014-09-18']
    elif (id1 == 542646) or (id2 == 542646):
        overlap_dates = ['2012-10-26'] 
    elif (id1 == 417635)  or (id2 == 417635):
        overlap_dates = ['2011-05-23','2011-05-24','2011-05-25','2011-05-26','2011-05-27','2011-05-28'] 
        
    # MAKE PLOT
    colors = index_cmap(len(overlap_dates),'plasma')
    for d in np.arange(0,len(overlap_dates)):
    
        contour_lons1,contour_lats1,center_lon1,center_lat1 = get_eddy_by_ID_date(ds,id1,overlap_dates[d])
        if not(all_equal(contour_lons1)): # if eddy was not detectable on a given day, it returns in AVISO as all a single coordinate
            plt.plot(contour_lons1,contour_lats1,c=colors[d])
        
        contour_lons2,contour_lats2,center_lon2,center_lat2 =get_eddy_by_ID_date(ds,id2,overlap_dates[d])
        if not(all_equal(contour_lons2)):
            plt.plot(contour_lons2,contour_lats2,c=colors[d],label=overlap_dates[d])
    
    plt.xlabel('Longitude ($^{\circ}$E)');
    plt.ylabel('Latitude ($^{\circ}$N)');
    plt.legend(bbox_to_anchor=(1,1));
    plt.show()
```
